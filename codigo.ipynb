```python
import pandas as pd
import numpy as np
```

```python
books = pd.read_csv('data/Books.csv', low_memory = False)
books.head(3)
```

```python
ratings = pd.read_csv('data/Ratings.csv')
#ratings.head(5)
```

```python
# Data Process

books = books[['ISBN', 'Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher', 'Image-URL-L']]
books.rename(columns = {'Book-Title':'TITLE','Book-Author':'AUTHOR','Year-Of-Publication':'YEAR-OF-PUBLICATION','Publisher':'PUBLISHER', 'Image-URL-L':'IMAGE-URL-L'}, inplace = True)
#books.head(5)
```

```python
ratings = ratings[['User-ID', 'ISBN', 'Book-Rating']]
ratings.rename(columns = {'User-ID':'USER-ID', 'Book-Rating':'RATING'}, inplace = True)
#ratings.head(5)
```

```python
books.dropna(inplace = True)
books.isna().sum()
```

```python
ratings.isna().sum()
```

```python
ratings['USER-ID'].value_counts()
```

```python
qt_ratings = ratings['USER-ID'].value_counts() > 200
x = qt_ratings[qt_ratings].index
x.shape
```

```python
x
```

```python
ratings.shape
```

```python
ratings = ratings[ratings['USER-ID'].isin(x)]
ratings.shape
```

```python
ratings.head()
```

```python
books.head()
```

```python
books.info()
ratings.info()
```

```python
books_and_ratings = ratings.merge(books, on = 'ISBN')
books_and_ratings.head()
```

```python
books_and_ratings.shape
```

```python
num_rating = books_and_ratings.groupby('TITLE')['RATING'].count().reset_index()
num_rating.rename(columns={'RATING' : 'NUM_OF_RATING'}, inplace=True)
num_rating.head(5)
```

```python
final_rating = books_and_ratings.merge(num_rating, on='TITLE')
```

```python
final_rating.head(2)
```

```python
final_rating = final_rating[final_rating['NUM_OF_RATING']>=50]
final_rating.sample(10)
```

```python
final_rating.drop_duplicates(['USER-ID', 'TITLE'], inplace = True)
final_rating.sample(20)
```

```python
books_pivot = final_rating.pivot_table(columns = 'USER-ID', index = 'TITLE', values = 'RATING')
#books_pivot.head(20)
```

```python
books_pivot.fillna(0, inplace = True)
books_pivot.head()
```

```python
from scipy.sparse import csr_matrix

books_sparse = csr_matrix(books_pivot)
type(books_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm = 'brute')
model.fit(books_sparse)
```

```python
distances, sugestions = model.kneighbors(books_pivot.iloc[237,:].values.reshape(1, -1), n_neighbors=6)
distances  
```

```python
sugestions
```

```python
for i in range(len(sugestions)):
    print(books_pivot.index[sugestions[i]])
```

```python
books_name = books_pivot.index
books_name
```

```python
def recomend_book(books_name):
    book_id = np.where(books_pivot.index == books_name)[0][0]
    distances, sugestions = model.kneighbors(books_pivot.iloc[book_id,:].values.reshape(1, -1), n_neighbors=6)

    for i in range(len(sugestions)):
        books = books_pivot.index[sugestions[i]]
        for j in books:
            print(j)
```
